```python
# Mô tả dữ liệu:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Đọc dữ liệu từ file csv
data = pd.read_csv("data.csv")
data.head() # Hiển thị 5 dòng đầu tiên của dữ liệu
```

```python
# Hiển thị thông tin về dữ liệu, bao gồm số lượng dòng, số lượng cột, và thông tin về kiểu dữ liệu của từng cột
data.info()
```

```python
# Hiển thị mô tả thống kê của dữ liệu, bao gồm các thông số như số lượng, trung bình, min, max, và phân vị của từng cột
data.describe()
```

```python
# Vẽ biểu đồ heatmap để kiểm tra dữ liệu thiếu (null) trong bảng dữ liệu
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Tạo ma trận tương quan của dữ liệu
corr_matrix = data.corr()

# Vẽ biểu đồ heatmap tương quan để xem mối quan hệ giữa các đặc tính dữ liệu
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Loại bỏ các cột không cần thiết khỏi dữ liệu
data.drop(['Unnamed: 32', "id"], axis=1, inplace=True)
```

```python
# Chuyển đổi nhãn của cột "diagnosis" từ chuỗi ("M" hoặc "B") sang số (1 hoặc 0)
data.diagnosis = [1 if value == "M" else 0 for value in data.diagnosis]
```

```python
# Chuyển đổi kiểu dữ liệu của cột "diagnosis" sang dạng categorical
data['diagnosis'] = data['diagnosis'].astype('category', copy=False)

# Vẽ biểu đồ cột để hiển thị phân phối của các lớp trong cột "diagnosis"
plot = data['diagnosis'].value_counts().plot(kind='bar', title="Class distributions \n(0: Benign | 1: Malignant)")
fig = plot.get_figure()
```

```python
# Xác định biến mục tiêu (nhãn)
y = data["diagnosis"]

# Xác định các biến dự đoán (các đặc tính khác trừ cột "diagnosis")
X = data.drop(["diagnosis"], axis=1)
```

```python
# Tiền xử lý dữ liệu bằng phương pháp chuẩn hóa dữ liệu
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Chia dữ liệu thành tập huấn luyện và tập kiểm tra
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.30, random_state=42)
```

```python
# Sử dụng mô hình hồi quy logistic
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

# Huấn luyện mô hình trên tập huấn luyện
lr.fit(X_train, y_train)

# Dự đoán biến mục tiêu trên tập kiểm tra
y_pred = lr.predict(X_test)
```

```python
# Đánh giá độ chính xác của mô hình
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
# Đánh giá chi tiết về hiệu suất của mô hình
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
